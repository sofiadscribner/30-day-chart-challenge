```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = {
    'activity': ['Sleeping (32%)', 'Hanging out with Husband (12%)', 'Class (16%)', 'Homework (13%)', 'Tutoring (21%)', 'Commute (6%)'],
    'percent': [0.3264, 0.125, 0.1563, 0.1285, 0.2083, 0.0555]
}

df = pd.DataFrame(data)

# sort categories from most to least time spent

df_sorted = df.sort_values(by='percent', ascending=False)

# initialize the chart

fig, ax = plt.subplots(figsize=(8, 8))
size = 0.3

# specify colors using one of seaborn's color palettes

colors = sns.color_palette("PuBu", n_colors=len(df_sorted))
colors = colors[::-1]

# plot the donut chart

ax.pie(df_sorted['percent'], labels=df_sorted['activity'], colors=colors, startangle=90, counterclock=False, 
       wedgeprops=dict(width=size, edgecolor='w'))
ax.set(aspect="equal")

# add the title in the center

ax.text(0, 0, 'A Glimpse into my Thursdays', ha='center', va='center', fontsize=14, fontweight='bold')

# save the plot locally
# plt.savefig('donut_chart_with_center_title_no_percentages.png', dpi=300, bbox_inches='tight')

plt.show()
```
